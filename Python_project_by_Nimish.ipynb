```python
# @title  Outlier
num_var=df.select_dtypes(exclude='object')
for i in num_var:
  q1=df[i].quantile(0.25)
  q3=df[i].quantile(0.75)
  IQR=q3-q1
  ul=q3+1.5*IQR
  ll=q1-1.5*IQR
  print(IQR,'IQR')
  print(ul,'ul')
  print(ll,'ll')
  print(i,df.shape[0]-df[(df[i]<ul) & (df[i]>ll)].shape[0])
  print('')
```
